```python
#now we will investigate the sqft column and find the outliers
df3.total_sqft.unique()
```

```python
#removing the range by taking the average of the range
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

df3[~df3["total_sqft"].apply(is_float)].head(10)
```

```python
#so here we can see a lot of errors and bad data in the dataset
# so for it we will be doing data cleaning

def convert_sqft_to_num(x):
    tokens = x.split(" - ")
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
#creating the new dataframe in the data processing pipeline

df4 = df3.copy()
df4["total_sqft"] = df4["total_sqft"].apply(convert_sqft_to_num)
df4.head()
```

```python
#feature engineering
df5 = df4.copy() #deep copy
# for outlier detection
df5["price_per_sqft"] = df5["price"]*100000/df5["total_sqft"]
df5.head()
```

```python
len(df5.location.unique()) #high dimensionality problem
```

```python
df5.location = df5.location.apply(lambda x: x.strip())
location_stats = df5.groupby("location")["location"].agg("count").sort_values(ascending = False)
location_stats
```

```python
len(location_stats[location_stats<= 10])
location_stats_less_than_10 = location_stats[location_stats <= 10]
location_stats_less_than_10
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: "other" if x in location_stats_less_than_10 else x)
```

```python
#so here we have made all the locations with less than 10 houses to others to reduce the dimensionality
#here feature engineering ends
```

```python
len(df5["location"] == "other")
```

```python
#outlier detection and removal
# we will detect the outlier by sqft per bedroom > 300sqft per bedroom
```

```python
# df5["sqft_per_bhk"] = df5.total_sqft/df5.bhk
df5[df5.total_sqft/df5.bhk < 300].head()
# df5 = df5.drop("sqft_per_bhk", axis = 1)
df5.head()
```

```python
#creating a new dataframe and removing the outliers
df6 = df5[~(df5.total_sqft/df5.bhk < 300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby("location"):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index = True)
    return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def plot_scatter_chart(df, location):
    bhk2 = df[(df.location == location) & (df.bhk == 2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]
    matplotlib.rcParams["figure.figsize"] = (6,9)
    plt.scatter(bhk2.total_sqft, bhk2.price,color="blue", label = "2 BHK", s = 50)
    plt.scatter(bhk3.total_sqft,bhk3.price, marker = "+", color = "green", label = "3 BHK", s = 50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()

plot_scatter_chart(df7, "Hebbal")
```

```python
def remove_bhk_outlier(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby("location"):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby("bhk"):
            bhk_stats[bhk] = {
                "mean": np.mean(bhk_df.price_per_sqft),
                "std": np.std(bhk_df.price_per_sqft),
                "count": bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby("bhk"):
            stats = bhk_stats.get(bhk-1)
            if stats and stats["count"] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats["mean"])].index.values)
    return df.drop(exclude_indices, axis = "index")

df8 = remove_bhk_outlier(df7)
df8.shape
```

```python
#finding the bathroom outliers by using histograms
# the number of bathrooms should not be more than bhk+2

# so all houses having bathrooms more than 2+bhk are outliers
```

```python
#removing the bathrooms outliers
df8[df8.bath>df8.bhk+2]
```

```python
df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
#finalizing the dataset and dropping the other columns
df10 = df9.drop(["size", "price_per_sqft"], axis = "columns")
df10.head()
```

```python
#so these are the 5 parameters that would be deciding the price of the real estate and we would be using these features for building the models
```

```python
#we have to convert the location to number
# so we apply the one hot encoding
# len(df10.location.unique()) -> number of unique locations
dummies = pd.get_dummies(df10.location)
dummies.head(10)
```

```python
df11 = pd.concat([df10, dummies.drop("other", axis = "columns")], axis = "columns")
df11.head()
```

```python
df12 = df11.drop("location", axis = "columns")
df12.head()
```

```python
df12.shape
```

```python
X = df12.drop("price", axis = "columns")
X.head()
```

```python
y = df12.price
y.head()
```

```python
#now applying the test train split method
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# now improving the score -> Crossvalidation
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
#shuffle split better shuffles the dataset than train test split so we will now compare both the models
cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)
cross_val_score(LinearRegression(), X, y, cv = cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearch(X,y):
    algos = {
        "linear_regression": {
            "model": LinearRegression(),
            "params": {
                "normalize": [True, False],
            }
        },
        "lasso": {
            "model": Lasso(),
            "params": {
                "alpha": [1,2],
                "selection": ["random", "cyclic"]
            }
        },
        "decision_tree": {
            "model": DecisionTreeRegressor(),
            "params": {
                "criterion": ["mse", "friedman_mse"],
                "splitter": ["best", "random"]
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config["model"], config["params"], cv =cv, return_train_score = False)
        gs.fit(X,y)
        scores.append({
            "model": algo_name,
            "best_score": gs.best_score_,
            "best_params": gs.best_params_
        })
    return pd.DataFrame(scores,columns=["model","best_score","best_params"])

find_best_model_using_gridsearch(X,y)
```

```python
#predicting the prices
def predict_price(location,sqft,bath,bhk):
    loc_index = np.where(X.columns == location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
